```python
import wget

wget.download(url='https://archive.ics.uci.edu/ml/machine-learning-databases/00312/dow_jones_index.zip', out='./sample_data/data.zip')
```

```python
import zipfile

with zipfile.ZipFile('./sample_data/data.zip', 'r') as fp:
    fp.extractall('./sample_data')
```

```python
import os 

os.remove('./sample_data/data.zip')
os.rename('./sample_data/dow_jones_index.data', './sample_data/dow_jones_index.csv')
```
